## **Fórmulas**

### **Grayscale**

Lightness: $\frac{max(R,G,B)+min(R,G,B)}{2}$

Average: $\frac{R+G+B}{3}$

Luminosity: $0.2126 \cdot R + 0.7152 \cdot G + 0.0722 \cdot B$


### **Cor**

$$\begin{bmatrix}
    R \\ G \\ B
\end{bmatrix} = \begin{bmatrix}
    255 \\ 255 \\ 255
\end{bmatrix} - \begin{bmatrix}
    C \\ M \\ Y
\end{bmatrix}$$

**b)** Faça a reflexão da imagem anterior nos eixos X e Y. Exiba o resultado.

````python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

farm = np.array(Image.open('farm.jpg'))[:, :, :3]
(l, c, p) = farm.shape
print(farm.shape)
plt.imshow(farm)

# Reflexão em y:
farm_refl = np.zeros(shape=farm.shape, dtype=np.uint8)
for i in range(l):
    for j in range(c):
        new_x = -j
        new_y = i
        farm_refl[new_y, new_x] = farm[i, j]

plt.imshow(farm_refl)
````

````python
# Reflexão em x:
farm_refl = np.zeros(shape=farm.shape, dtype=np.uint8)
# Acesso a índices: começo:fim:passo
# Logo: ::-1 significa do começo ao fim com passo -1 (de trás para frente)
farm_refl[::-1, ::1] = farm # ou mario[:, :]

plt.imshow(farm_refl)
````

````python
#NOVO METODO

# Abre a imagem
imagem = Image.open("farm.jpg")

# Cria uma cópia da imagem
nova_imagem = imagem.copy()

# Obtém os pixels da imagem
pixels = nova_imagem.load()

# Inverte a imagem em x e y
for i in range(nova_imagem.width):
    for j in range(nova_imagem.height):
        pixels[i, j] = pixels[nova_imagem.width - 1 - i, nova_imagem.height - 1 - j]

# Salva a nova imagem
nova_imagem.save("nova_imagem.jpg")
plt.imshow(nova_imagem)
````

**c)** Aumente o brilho da imagem da questão anterior em 1,6. Exiba o resultado.

````python
# Abre a imagem
imagem = Image.open("farm.jpg")

# Cria uma cópia da imagem
nova_imagem = imagem.copy()

# Obtém os pixels da imagem
pixels = nova_imagem.load()

# Percorre cada pixel da imagem e aumenta o brilho em 1.6x
for i in range(nova_imagem.width):
    for j in range(nova_imagem.height):
        r, g, b = pixels[i, j]
        r = int(r * 1.6)
        g = int(g * 1.6)
        b = int(b * 1.6)
        pixels[i, j] = (r, g, b)

# Salva a nova imagem
nova_imagem.save("farmC.jpg")
plt.imshow(nova_imagem)
````

**e)** Realize uma operação de blending entre a imagem da questão anterior e a imagem da questão **C**. Utilize um coeficiente de 20% para a imagem da questão **C**. Exiba o resultado.

````python

def blend(img1, img2, c):
    return (img1 * c + img2 * (1 - c)).astype(np.uint8)

spaceship = np.array(Image.open('spaceship.jpg'))
farmC = np.array(Image.open('farmC.jpg'))
blended = blend(spaceship, farmC, 0.2)

blended.save('blended.jpg')
plt.imshow(blended)
````

**f)** Gere o histograma da imagem anterior nos três canais de cor e apresente o resultado.

````python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
img = np.array(Image.open('spaceship.jpg'))[:, :, :3]
plt.figure(figsize=(8, 8))
plt.imshow(img)

def lightness(img):
    M = np.max(img, axis=2).astype(np.float64)
    m = np.min(img, axis=2).astype(np.float64)
    return ((M + m) / 2).astype(img.dtype)

def average(img):
    r = img[:, :, 0].astype(np.float64)
    g = img[:, :, 1].astype(np.float64)
    b = img[:, :, 2].astype(np.float64)
    return ((r + g + b) / 3).astype(img.dtype)

def luminosity(img):
    r = img[:, :, 0].astype(np.float64)
    g = img[:, :, 1].astype(np.float64)
    b = img[:, :, 2].astype(np.float64)
    return (0.2126 * r + 0.7152 * g + 0.0722 * b).astype(img.dtype)
````
