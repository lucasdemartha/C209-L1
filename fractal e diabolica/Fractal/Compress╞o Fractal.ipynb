# [Compresso fractal](https://github.com/pvigier/fractal-image-compression)

* Baseia-se em fractais;
* Parte do fato que partes de uma imagem assemelham-se a outras partes;
* Adequado para comprimir texturas e imagens naturais;
* Computacionalmente custoso devido  busca de similaridades.



#### Conceitos bsicos

* Uma *funo afim*  uma funo polinomial de 1 grau consistindo de uma transformao linear e uma translao em R (exemplos: espelhamento e rotao);
* Usaremos apenas funo afins contrativas. Uma funo afim  dita contrativa se seu fator contrativo `s`  menor que `1` (i.e., ele "reduz" o tamanho das imagens);
* Dada uma coleo finita de mapas contrativos afins $W_1, W_2, \dots, W_n$, essa coleo forma um sistema de funes iteradas (*iterated functions system*, IFS).

#### Compresso

* Observou-se que imagens naturais frequentemente so ricas em *redundncias afins*. Isso significa que com um IFS adequado, partes maiores de uma imagem so similares a partes menores;
* Primeiro, divide-se a imagem em parties (de qualquer formato e tamanho) chamadas *regies de domnio*;
* Em seguida, uma coleo de possveis *regies de alcance*  definida;
    * Gera-se todas as regies de alcance possveis para todas as regies de domnio;
* Para cada regio de domnio, o algoritmo busca uma regio de alcance adequada  que, quando aplicada com uma funo afim apropriada, se assemelha muito  regio de domnio;
    * Cada regio de alcance  testada, e salva-se a mais prxima da partio original.
* Por fim, um formato de imagem fractal (*Fractal Image File*, FIF)  gerado, contendo as escolhas de regies de domnio e a lista de coeficientes afins de tidas as funes afins associadas.

#### Descompresso

* Para descomprimir, primeiro a imagem  particionada em regies de domnios como especificadas no FIF;
* Para cada regio de domnio, a sua regio de alcance associada  encontrada, e a funo afim correspondente  aplicada, convergindo a imagem para o atrator do mapa;
* O processo  repetido vrias vezes. Quando mais iteraes forem aplicadas, mais prxima a imagem fica semelhante  original.
