# [Compressão fractal](https://github.com/pvigier/fractal-image-compression)

* Baseia-se em fractais;
* Parte do fato que partes de uma imagem assemelham-se a outras partes;
* Adequado para comprimir texturas e imagens naturais;
* Computacionalmente custoso devido à busca de similaridades.



#### Conceitos básicos

* Uma *função afim* é uma função polinomial de 1º grau consistindo de uma transformação linear e uma translação em Rⁿ (exemplos: espelhamento e rotação);
* Usaremos apenas função afins contrativas. Uma função afim é dita contrativa se seu fator contrativo `s` é menor que `1` (i.e., ele "reduz" o tamanho das imagens);
* Dada uma coleção finita de mapas contrativos afins $W_1, W_2, \dots, W_n$, essa coleção forma um sistema de funções iteradas (*iterated functions system*, IFS).

#### Compressão

* Observou-se que imagens naturais frequentemente são ricas em *redundâncias afins*. Isso significa que com um IFS adequado, partes maiores de uma imagem são similares a partes menores;
* Primeiro, divide-se a imagem em partições (de qualquer formato e tamanho) chamadas *regiões de domínio*;
* Em seguida, uma coleção de possíveis *regiões de alcance* é definida;
    * Gera-se todas as regiões de alcance possíveis para todas as regiões de domínio;
* Para cada região de domínio, o algoritmo busca uma região de alcance adequada  que, quando aplicada com uma função afim apropriada, se assemelha muito à região de domínio;
    * Cada região de alcance é testada, e salva-se a mais próxima da partição original.
* Por fim, um formato de imagem fractal (*Fractal Image File*, FIF) é gerado, contendo as escolhas de regiões de domínio e a lista de coeficientes afins de tidas as funções afins associadas.

#### Descompressão

* Para descomprimir, primeiro a imagem é particionada em regiões de domínios como especificadas no FIF;
* Para cada região de domínio, a sua região de alcance associada é encontrada, e a função afim correspondente é aplicada, convergindo a imagem para o atrator do mapa;
* O processo é repetido várias vezes. Quando mais iterações forem aplicadas, mais próxima a imagem fica semelhante à original.
