### Viso geral do algoritmo

1. Em blocos quadrados, calcula-se a mdia dos pixels;
2. Determina-se o pixel que tem a menor distncia de cor em relao  mdia;
3. Atribui-se o pixel determinado para o bloco inteiro;
4. Aplica-se a compresso RLE no resultado.

### Distncia de cor
* Obtem-se uma diferena numrica que baseia-se na percepo visual humana.

$$\Delta E_{ab}^*=\sqrt{(L_2^*-L_1^*)^2+(a_2^*-a_1^*)^2+(b_2^*-b_1^*)^2}$$

No qual $(L^*, a^*, b^*)$ [esto no espao de cor LAB](https://en.wikipedia.org/wiki/CIELAB_color_space).
