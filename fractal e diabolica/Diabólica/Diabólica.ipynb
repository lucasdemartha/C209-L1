* Deterministic Iterative Average Block-Oriented Lossy Image Compression Algorithm;
* Compressão de imagens;
* Diferentemente de outros métodos, não utiliza transformada;
* Trabalha com blocos quadrados de pixels para gerar áreas homogêneas na imagem.

### Visão geral do algoritmo

1. Em blocos quadrados, calcula-se a média dos pixels;
2. Determina-se o pixel que tem a menor distância de cor em relação à média;
3. Atribui-se o pixel determinado para o bloco inteiro;
4. Aplica-se a compressão RLE no resultado.

### Distância de cor
* Obtem-se uma diferença numérica que baseia-se na percepção visual humana.

$$\Delta E_{ab}^*=\sqrt{(L_2^*-L_1^*)^2+(a_2^*-a_1^*)^2+(b_2^*-b_1^*)^2}$$

No qual $(L^*, a^*, b^*)$ [estão no espaço de cor LAB](https://en.wikipedia.org/wiki/CIELAB_color_space).
